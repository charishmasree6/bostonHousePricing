```python
# 1. Start with the feature data (already a DataFrame from fetch_openml)
dataset = boston_df.data.copy() 

# 2. Add the target column and explicitly name it 'Price'
# The target data is in the 'target' attribute of the boston_df object
dataset['Price'] = boston_df.target

# 3. Ensure the target column is the correct numerical type (float)
dataset['Price'] = dataset['Price'].astype(float) 

# Verify the fix: check the columns
print("Columns in dataset:", dataset.columns.tolist())
```

```python
# Separate features (all columns except 'Price') into x
x = dataset.drop('Price', axis=1)

# Separate the target column ('Price') into y
y = dataset['Price']
```

```python
#Model Training
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
#R square and adjusted R square
#R^2=1-ssr/sst 
#adjusted R^2=1-[(1-R2)*(n-1)/(n-k-1)]
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
# display adjusted R-squared
1 - (1 - score) * (len(y_test) - 1) / (len(y_test) - x_test.shape[1] - 1)
```
